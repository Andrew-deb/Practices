```python
people = {
        "first": ["Corey", "Jane", "John"],
        "last":  ["schafer", 'Doe', 'Doe'],
        "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@gmail.com', 'JohnDoe@gmail.com']
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
 filt = (df['last'] == 'Doe')
```

```python
df[filt]
```

```python
df.loc[filt] #.loc() used to look up rows and columns by labels
```

```python
#We could also pass in more argument into the .loc() function to look up specific columns
df.loc[filt, 'email']
```

```python
#To carryout filtering operations on pandas we could use the built-in boolean functions in python
#But this time: The "and" boolean is represented by &
#And the "or" boolean is represented by |
```

```python
#For example
filt = ((df['last'] == 'Doe') & (df["first"]== 'John'))
```

```python
df.loc[filt]
```

```python
filt= ((df['last']== 'schafer') | (df['first']== 'John'))
```

```python
df.loc[filt]
```

```python
#To get the negation of any boolean operation carried out, simply place a tilda(-) sign before the whole operation or the varaible the opreation is saved in
#For example:
df.loc[-filt] #Which reads fetch the rows which the last name is not schafer and the first name is not John
```
