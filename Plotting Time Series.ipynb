```python
import pandas as pd
from matplotlib import dates as mpl_dates
from matplotlib import pyplot as plt
from datetime import datetime, timedelta
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
dates= [
        datetime(2019, 5, 24),
        datetime(2019, 5, 25),
        datetime(2019, 5, 26),
        datetime(2019, 5, 27),
        datetime(2019, 5, 28),
        datetime(2019, 5, 29),
        datetime(2019, 5, 30)
      ]
y= [0, 1, 3, 4, 6, 5, 7]
```

```python
#This command plots our y-axis points to our x-axis
plt.plot_date(dates, y)
plt.tight_layout()
plt.show()
```

```python
#We could also pass the linestyle argument to connect our points together
plt.plot_date(dates, y, linestyle= 'solid')
plt.tight_layout()
plt.show()
```
