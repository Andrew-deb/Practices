```python
#Get the percentage income of every developers in each country, showing which country pays their developers the highest.
df4= df[['Country', 'ConvertedComp']]
```

```python
df4.isna()
```

```python
df4.dropna(axis= 'index', how= 'any', inplace= True)
```

```python
df4.head()
```

```python
country_grp= df4.groupby(['Country'])
```

```python
country_grp.get_group('United States')
```

```python
df4.rename(columns={'ConvertedComp': 'SalaryUSD'}, inplace= True)
```

```python
df5= country_grp['SalaryUSD'].agg(['count', 'median'])
```

```python
df5.reset_index(inplace=True)
```

```python
df5.head()
```

```python
df5.rename(columns={'count': 'total_num_dev', 'sum': 'total_salaries', 'median': 'Avr_dev_sal'}, inplace= True)
```

```python
df5
```

```python
df5['Avr_dev_sal']= df5['Avr_dev_sal'].astype('float')
```

```python
df5.to_csv('data/salaries.csv')
```

```python
def per_sal(x):
    sum= 0
    for i in df5['Avr_dev_sal']:
        sum= sum+i
    return round(((x/sum) *100), 2)
```

```python
df5['% dev_sal']= df5['Avr_dev_sal'].apply(per_sal)
```

```python
from collections import Counter
import csv
job_factor= Counter()
```

```python
with open('data/job_fac.csv') as csv_file:
    csv_reader= csv.DictReader(csv_file)
    
    job_fac_counter= Counter()
    for row in csv_reader:
        job_fac_counter.update((row['WorkChallenge'].split(';')))
print(job_fac_counter.most_common(9))

Work_cha= []
Devs= []
```

```python
Work_cha= []
Devs= []
for items in job_fac_counter.most_common(9):
    Work_cha.append(items[0])
    Devs.append(items[1])
print(Work_cha)
print(Devs)
```

```python
Work_cha[1]= 'Non-work commitments'
```

```python
from matplotlib import pyplot as plt
```

```python
#plt.style.use('seaborn-v0_8-notebook')
Work_cha.reverse()
Devs.reverse()
plt.barh(Work_cha, Devs, color= 'c')
plt.xlabel('Number of developers that faces this challenge')
plt.title('Most common Challenges amongst high earning developers')
plt.tight_layout()
plt.savefig('C:/Users/USER/Desktop/Data Analytics materials/Matplotlib/Analytical_survey.png')
plt.show()
```
