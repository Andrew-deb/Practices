```python
import pandas as pd
```

```python
df= pd.read_csv('data/survey_results_public.csv', index_col= 'Respondent')
schema_df=  pd.read_csv('data/survey_results_schema.csv', index_col= 'Column')
```

```python
pd.set_option('display.max_columns', 85)
pd.set_option('display.max_rows', 85)
```

```python
df.head()
```

```python
sample_df= df[['Age', 'LanguageWorkedWith', 'ConvertedComp']]
```

```python
sample_df.dropna(axis= 'index', how= 'any', inplace= True)
```

```python
sample_df.head()
```

```python
filt= (sample_df['Age'] >= 18) & (sample_df['Age'] <= 37)
```

```python
sample_df= sample_df.loc[filt,['Age','LanguageWorkedWith', 'ConvertedComp']]
```

```python
sample_df
```

```python
Age_grp= sample_df.groupby('Age')
```

```python
#Age_grp['ConvertedComp'].agg('median')
median_salary_by_age= Age_grp['ConvertedComp'].median()
median_salary_by_age
```

```python
python_users_by_age= Age_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
python_users_by_age.head()
```

```python
python_users_df= pd.DataFrame({'MedianSalary': median_salary_by_age, 'PythonUsers': python_users_by_age})
```

```python
python_users_df.head()
```

```python
python_users_df = python_users_df[python_users_df['PythonUsers'] > 0]
```

```python
python_users_df
```

```python
Js_users_by_age= Age_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('JavaScript').sum())
Js_users_df= pd.DataFrame({'MedianSalary': median_salary_by_age, 'JavaScript_Users': Js_users_by_age})
Js_users_df = Js_users_df[Js_users_df['JavaScript_Users'] > 0]
Js_users_df
```

```python
result_df= { }
```

```python
result_df= pd.DataFrame(result_df)
```

```python
result_df['All_dev']= median_salary_by_age
```

```python
result_df['Python'] = python_users_df['MedianSalary']
```

```python
result_df['JavaScript']= Js_users_df['MedianSalary']
```

```python
result_df.dropna(axis= 'index', how= 'any', inplace=True)
```
