```python
import pandas as pd
import numpy as np
```

```python
people= {
        'first': ['Corey', 'Jane','John', 'Chris', np.nan, None, 'NA'],
        'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'],
        'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@gmail.com', 'JohnDoe@gmail.com', None, np.nan, 'Anonymous@gmail.com', 'NA'],
        'age': ['33', '55', '63', '36', None, None, 'Missing']
}
```

```python
df= pd.DataFrame(people)
```

```python
df
```

```python
#From our dataframe we could observe that they are some missing values indicated by nan or None
#To drop rows of columns with missing values

df.dropna()
```

```python
#Note by default the .drop() function drops any rows with missing values and it could also be expressed as:
#df.dropna(axis= 'index', how= 'any') #This command only drops rows when any the values are missing
```

```python
#But in cases where we seek to remove rows only when the all its columns are missing ie reverse a row when it has some missing value

df.dropna(axis='index', how= 'all') #This command only drops rows when all the values are missing
```

```python
#If we want to remove by column
df.dropna(axis= 'columns', how= 'all')
```

```python
#You could also drop by a particular row or column pending on which element in that row or column is a nan or None value
df.dropna(axis='index', how='all', subset= ['email'])
```

```python
#The subset command could also contain two or more rows
df.dropna(axis='index', how='all', subset= ['email', 'last']) #This line of command is only allowed to drop any row that doesn't have a last name or an email
```

```python
#We could replace our custom missing values in our dataframe to nan values, using the replace command
df.replace('NA', np.nan, inplace= True)
df.replace('Missing', np.nan, inplace= True)
```

```python
df
```

```python
#We could check to confirm all the columns that hold as nan value in our dataframe with the isna() command
df.isna()
```

```python
df.dtypes #To get the datatypes of each column
#Note that the object datatype signifies that the column is of string type or has mixture of various datatypes.
```

```python
#Since it is an object datatype to perform mathematical operation on the age column we need to convert string datatype to float since the nan values are already of float type
type(np.nan)
```

```python
#Note that converting columns with nan values to integer type would result to an error, because the nan values are constantly of the float type
#df['age'] = df['age'].astype(int)
```

```python
df['age'] = df['age'].astype(float)
```

```python
df.dtypes #Now our age column is of float type
```

```python
df['age'].mean() #Computes the mean for the age column while stile maintaining the nan values as missing values 
```

```python
#Performing cleaning and casting operations on the YearsCode column 
df['YearsCode']= df['YearsCode'].astype('float') #Since it contains nan values
```

```python
#Our code resulted to an error because we had string datatypes that could not be converted to float and we could view those values using the unique method
df['YearsCode'].unique()
```

```python
#We can see that we have two strings in our YearsCode column. We could resolve the error by replacing the string values with an integer or float
df['YearsCode'].replace('Less than 1 year', 0, inplace= True) #Here we are setting our less than a year string value to 0 since the years code is not up to a year
```

```python
df['YearsCode'].replace('More than 50 years', 51, inplace= True) #Here we assume a value of 51 for the string value of more than a year
```

```python
df['YearsCode'].unique()
```

```python
#Now we could convert the elements in the column into float types
df['YearsCode']= df['YearsCode'].astype('float')
```

```python
#Now we could perform statistical operation on the YearsCode column
df['YearsCode'].mean() 
```
